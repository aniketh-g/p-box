Use these functions to generate lists of minterms and don't cares. Feed these into a tool like https://www.charlie-coleman.com/experiments/kmap/ and generate the logic expression

```python
def complement(univ_list, spec_list):
    return [i for i in univ_list if i not in spec_list]
def getDCs(ones, zeroes):
    return [i for i in list(range(128)) if i not in ones and i not in zeroes]
def genKmpIPs(super_set_ones, subset_ones):
    subset_zeroes = complement(super_set_ones, subset_ones)
    dont_cares = getDCs(subset_ones, subset_zeroes)
    print('ones : ', *subset_ones, sep=',')
    print('zeros: ', *subset_zeroes, sep=',')
    print('DCs  : ', *dont_cares, sep=',')
    print('')
    return [subset_ones, subset_zeroes, dont_cares]
```

```python
dcs = [4,5,6,7,13,14,21,22,23,30,36,47,55,59,63,67,74,75,82,83]
for i in range(8,16):
    dcs.append(8*i+4)
    dcs.append(8*i+5)
    dcs.append(8*i+6)
    dcs.append(8*i+7)
total = complement(list(range(128)), dcs)
print(dcs)
```

```python
print('is32BitMulAcc')
is32BitMulAcc = genKmpIPs(total, [12, 20, 28, 29, 37, 38, 39, 44, 45, 46, 52, 53, 54, 60, 61, 62])
# (f7[2] & ~f7[1]) | (f7[5] & f7[2]);

# Pre Mul:
print('isSwitchB')
isSwitchB = genKmpIPs(is32BitMulAcc[0], [12,20,29,37,39,53,60,61,62])
# (~f7[5] & ~f7[4]) | (~f7[5] & ~f7[3]) | (f7[5] & f7[4] & f7[3]) | (~f7[5] & f7[0]) | (~f7[3] & f7[0]);

print('isSwitchT')
isSwitchT = genKmpIPs(isSwitchB[0], [20,61])
# (f7[5] & f7[3] & f7[0]) | (~f7[5] & f7[4] & ~f7[0]);
```
